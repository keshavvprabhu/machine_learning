## numpy tutorial

```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
## Create an array of integers
a1 = np.array([3,4.5,4.8,6])
a1
plt.plot(list(range(len(a1))),a1 )
```

```python
## Create an array of 10 Zeros
a2 = np.zeros(10)
a2
```

```python
## Create an array of 10 Ones
a3 = np.ones(10)
a3
```

```python
## Create an array of 10 randome numbers
a4 = np.random.random(10)
a4
plt.plot(list(range(len(a4))), a4 )
```

```python
"""
If positive int_like arguments are provided, randn generates an array of shape (d0, d1, ..., dn), filled with random floats sampled from a univariate “normal” (Gaussian) distribution of mean 0 and variance 1
"""
a5 = np.random.randn(10)
print(a5)
plt.plot(list(range(len(a5))), a5 )
```

```python
a6 = np.linspace(0,10, 100)
print(a6)
plt.plot(list(range(len(a6))), a6 )
```

```python
a7 = np.arange(0,10, 0.25)
print(a7)
plt.plot(list(range(len(a7))), a7 )
```

# Array Operations


```python
print(a1)
print(a1 * 2)
print(a1 > 3)
```

```python
x = np.arange(-1000, 1000, 1)
x
for i in range(0,7):
  y = x**i
  print(f"Degree : {i}")
  plt.plot(x, y)
  time.sleep(2)
  plt.show()
  
```
