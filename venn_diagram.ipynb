```python
#Import libraries
from matplotlib_venn import venn2, venn2_circles, venn2_unweighted
from matplotlib import pyplot as plt
%matplotlib inline
missing_in_source = 30
missing_in_target = 30
mismatches = 5
v=venn2_unweighted(subsets = (missing_in_source, missing_in_target, mismatches), set_labels = ('Missing In Target', 'Missing In Source', 'Mismatches'), alpha=0.7, set_colors=('blue', 'green'))
v.get_patch_by_id('11').set_color('red')
plt.annotate('Mismatching Record Count', xy=v.get_label_by_id('11').get_position() - np.array([0, 0.05]), xytext=(-70,-140),
             ha='center', textcoords='offset points', bbox=dict(boxstyle='round, pad=0.5', fc='black', alpha=0.1),
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.4',color='black'))
plt.title("Data Comparison Summary - Mismatches")


```

```python
ax = plt.gca()
y = [104360,104360]
plt.barh(['Source_Record_Count', 'Target_Record_Count'],[104360,104360], alpha=0.7,
         color=("blue", "green"),height=0.6,  )
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
# ax.spines['left'].set_visible(False)
for i, v in enumerate(y):
    ax.text(v + 3, i, str(v), color = 'black', fontweight = 'bold')
plt.title("Data Comparison Summary - Source vs Target")
# plt.tight_layout()  
plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np
from matplotlib_venn import venn3, venn3_circles
plt.figure(figsize=(4,4))
v = venn3(subsets=(1, 1, 1, 1, 1, 1, 1), set_labels = ('A', 'B', 'C'))
v.get_patch_by_id('100').set_alpha(1.0)
v.get_patch_by_id('100').set_color('white')
v.get_label_by_id('100').set_text('Unknown')
v.get_label_by_id('A').set_text('Set "A"')
c = venn3_circles(subsets=(1, 1, 1, 1, 1, 1, 1), linestyle='dashed')
c[0].set_lw(1.0)
c[0].set_ls('dotted')
plt.title("Sample Venn diagram")
plt.annotate('Unknown set', xy=v.get_label_by_id('100').get_position() - np.array([0, 0.05]), xytext=(-70,-70),
             ha='center', textcoords='offset points', bbox=dict(boxstyle='round,pad=0.5', fc='gray', alpha=0.1),
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.5',color='gray'))
plt.show()
```
